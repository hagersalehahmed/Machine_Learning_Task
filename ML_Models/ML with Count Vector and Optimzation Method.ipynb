```python

#Create bag of word with n-gram between 3 and 7 using CountVectorizer
cv = CountVectorizer(analyzer = 'word',ngram_range=(3,7))
cv_train=cv.fit_transform(X_train['text'])
cv_test=cv.transform(X_test['text'])
```

```python
#print result
def printResult(y_real, y_pred):
    acc = accuracy_score(y_real, y_pred)
    print("Accuracy: {:.2f}".format(acc*100),end='\n\n')
    print(classification_report(y_real,y_pred))
```

**I used StratifiedKFold to select row from each class**


```python
 SFold = StratifiedKFold(n_splits=10, random_state=30, shuffle=True)
```

**Grid search with stratified cross validation to optimize model and select the best values **

```python
#Optimization Function
def Optimization_Model(model,model_Name, X_train,y_train, param_grid, SFold):
 
  grid_search = GridSearchCV(estimator = model, param_grid = param_grid, cv = SFold)
  grid_search.fit(X_train, y_train)
  grid_search.best_params_
  best_grid = grid_search.best_estimator_
  print(best_grid)
  return best_grid
```

**KNN is used to train and optmize train data

```python
KNN= KNeighborsClassifier()
param_grid = {'n_neighbors':[100,200,300]}

best_model_KNN=Optimization_Model(KNN,"KNN", cv_train,y_train, param_grid , SFold)
```

```python
model_KNN=best_model_KNN.fit(cv_train,y_train)
```
