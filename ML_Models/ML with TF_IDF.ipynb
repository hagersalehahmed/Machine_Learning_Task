##Loading DataSet
>load training dataset to train model and testing dataset to evaluate model
>I split dataset into two file traing set and testing set

```python
#Read train data
train= pd.read_csv('/content/drive/MyDrive/ML task/ML/train1.csv',encoding='utf-8')
train.dialect.value_counts()
```

```python
len(train)
```

```python
train.head()
```

```python
train=train.dropna()
```

```python
#train=train[0:2000]
```

```python
y_train = train.dialect                           
X_train= train.drop('dialect',axis = 1 )
```

```python
test= pd.read_csv("/content/drive/MyDrive/ML task/ML/unseen1.csv")
test=test.dropna()
y_test = test.dialect                           
X_test= test.drop('dialect',axis = 1 )

print(test['dialect'].value_counts())
```

```python

labelencoder_y = LabelEncoder()
y_train = labelencoder_y.fit_transform(y_train)
y_test = labelencoder_y.transform(y_test)
#len(y_train)
```

```python

#Create bag of word with n-gram between 3 and 7 usingTfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer( ngram_range=(3,7))

tfidf_train=tfidf.fit_transform(X_train['text'])
tfidf_test=tfidf.transform(X_test['text'])
print(tfidf_train.shape, tfidf_test.shape)
```

```python
joblib.dump(tfidf, "/content/drive/MyDrive/ML task/ML/TF_IDF/tfidf.pkl")
```

```python
#print result
def printResult(y_real, y_pred):
    acc = accuracy_score(y_real, y_pred)
    print("Accuracy: {:.2f}".format(acc*100),end='\n\n')
    print(classification_report(y_real,y_pred))
```

```python
KNN= KNeighborsClassifier()
model_KNN=KNN.fit(tfidf_train,y_train)
```

```python
y_test_pred=model_KNN.predict(tfidf_test)
print("Test Result of KNN")
printResult(y_test,y_test_pred)
```
